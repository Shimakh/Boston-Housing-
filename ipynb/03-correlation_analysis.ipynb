```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pandas.read_csv(url, names=names, sep='\s+')
#dataset.drop('MEDV', axis=1, inplace=True )
dataset.head(3)
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
def calculate_r_2_for_feature(data,feature):
    new_data = data.drop(feature, axis=1)

    X_train, \
    X_test,  \
    y_train, \
    y_test = train_test_split(
        new_data,data[feature],test_size=0.25
    )

    regressor = KNeighborsRegressor()
    regressor.fit(X_train,y_train)

    score = regressor.score(X_test,y_test)
    return score
```

```python
calculate_r_2_for_feature(dataset,'')
```

```python
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
print("{:24} {}".format(": ", calculate_r_2_for_feature(dataset,'')))
```

```python
def mean_r2_for_feature(data, feature):
    scores = []
    for _ in range(100):
        scores.append(calculate_r_2_for_feature(data, feature))
        
    scores = np.array(scores)
    return scores.mean()
```
