```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st

from IPython.display import display
%matplotlib inline
```

```python
import pandas
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pandas.read_csv(url, names=names, sep='\s+')
dataset.drop('CHAS', axis=1, inplace=True )
dataset.drop('ZN', axis=1, inplace=True )
dataset.head(3)
```

```python
data_samp_set = dataset.sample(5)
```

```python
samp_stats = data_samp_set.describe().T
samp_stats['skew'] = st.skew(data_samp_set)
samp_stats['kurt'] = st.kurtosis(data_samp_set)
samp_stats
```

```python
stats = dataset.describe().T
stats['skew'] = st.skew(dataset)
stats['kurt'] = st.kurtosis(dataset)
stats
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_sc = scaler.fit_transform(dataset)
data_sc_set = pd.DataFrame(data_sc, columns=dataset.columns)

sc_stats = dataset.describe().T
sc_stats['skew'] = st.skew(dataset)
sc_stats['kurt'] = st.kurtosis(dataset)
display(stats)
display(sc_stats)
```

```python
fig = plt.figure(figsize=(20,6))
for i, col in enumerate(dataset.columns):
    fig.add_subplot(4,4,1+i)
    sns.distplot(dataset[col], label=col)
    plt.axvline(dataset[col].mean(), c='red')
    plt.axvline(dataset[col].median(), c='black')
    plt.legend()
```

```python
import scipy.stats as st
```

```python
for col in data_sc_set.columns:
    original_col_skewtest = st.skewtest(dataset[col])
    scaled_col_skewtest = st.skewtest(data_sc_set[col])
    print("{}\norig skew test: {} \nscaled skew test: {}\n\n".format(col, 
                                                                     original_col_skewtest,
                                                                     scaled_col_skewtest))
```

```python
data_log_set = np.log(dataset)
scaler.fit(data_log_set)
data_log_sc = scaler.transform(data_log_set)
data_log_sc_set = pd.DataFrame(data_log_sc, columns=dataset.columns)
```

```python
for col in data_log_set.columns:
    original_col_skewtest = st.skewtest(dataset[col])
    scaled_col_skewtest = st.skewtest(data_sc_set[col])
    original_log_col_skewtest = st.skewtest(data_log_set[col])
    scaled_log_col_skewtest = st.skewtest(data_log_sc_set[col])
    print("""{}
    orig:       {} 
    scaled:     {}
    orig log:   {}
    scaled log: {}
    
    """.format(col, 
               original_col_skewtest,
               scaled_col_skewtest, 
               original_log_col_skewtest,
               scaled_log_col_skewtest)) 
```

```python
fig = plt.figure(figsize=(20,6))
for i, col in enumerate(data_sc_set.columns):
    fig.add_subplot(4,4,1+i)
    sns.distplot(data_sc_set[col], label=col)
    plt.axvline(data_sc_set[col].mean(), c='red')
    plt.axvline(data_sc_set[col].median(), c='black')
    plt.legend() 
```

```python
fig = plt.figure(figsize=(20,6))
for i, col in enumerate(data_log_sc_set.columns):
    fig.add_subplot(4,4,1+i)
    sns.distplot(data_log_sc_set[col], label=col)
    plt.axvline(data_log_sc_set[col].mean(), c='red')
    plt.axvline(data_log_sc_set[col].median(), c='black')
    plt.legend()
```

```python
data_log_sc_set.to_pickle('final_log_sc.p')
```

```python
data_box_cox_set = pd.DataFrame()
for col in dataset.columns:
    box_cox_trans = st.boxcox(dataset[col])[0]
    data_box_cox_set[col] = pd.Series(box_cox_trans)
```

```python
scaler.fit(data_box_cox_set)
data_box_cox_sc = scaler.transform(data_box_cox_set)
data_box_cox_sc_set = pd.DataFrame(data_box_cox_set, columns=dataset.columns)
```

```python
for col in data_log_set.columns:
    original_col_skewtest = st.skewtest(dataset[col])
    scaled_col_skewtest = st.skewtest(data_sc_set[col])
    original_log_col_skewtest = st.skewtest(data_log_set[col])
    scaled_log_col_skewtest = st.skewtest(data_log_sc_set[col])
    original_box_cox_col_skewtest = st.skewtest(data_box_cox_set[col])
    scaled_box_cox_col_skewtest = st.skewtest(data_box_cox_sc_set[col])
    print("""{}
    orig:           {} 
    scaled:         {}
    orig log:       {}
    scaled log:     {}
    orig box-cox:   {}
    scaled box-cox: {}
    
    """.format(col, 
               original_col_skewtest,
               scaled_col_skewtest, 
               original_log_col_skewtest,
               scaled_log_col_skewtest, 
               original_box_cox_col_skewtest,
               scaled_box_cox_col_skewtest))
```

```python
fig = plt.figure(figsize=(20,6))
for i, col in enumerate(data_box_cox_sc_set.columns):
    fig.add_subplot(4,4,1+i)
    sns.distplot(data_box_cox_sc_set[col], label=col)
    plt.axvline(data_box_cox_sc_set[col].mean(), c='red')
    plt.axvline(data_box_cox_sc_set[col].median(), c='black')
    plt.legend()
```

```python
data_box_cox_sc_set.to_pickle('final_box_cox_sc.p')
```
